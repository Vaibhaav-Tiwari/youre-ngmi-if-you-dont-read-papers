```python
import bs4
from langchain import hub
```

```python
from langchain.chains import create_retrieval_chain
```

```python
from langchain.chains.combine_documents import create_stuff_documents_chain
```

```python
from langchain_chroma import Chroma
```

```python
from langchain_community.document_loaders import WebBaseLoader
```

```python
from langchain_core.prompts import ChatPromptTemplate
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
loader = WebBaseLoader(
    web_paths=("https://lilianweng.github.io/posts/2023-06-23-agent/",),
    bs_kwargs=dict(parse_only=bs4.SoupStrainer(class_=("post-content", "post-title", "post-header"))),

)
```

```python
doc = loader.load()
```

```python
doc
```

```python
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=200,
)
```

```python
splits = text_splitter.split_documents(doc)
```

```python
splits
```

```python
from langchain_google_genai import GoogleGenerativeAIEmbeddings

gemini_embeddings = GoogleGenerativeAIEmbeddings(
    model="models/embedding-001",  # This must be exactly like this
    google_api_key="AIzaSyCs1veEtwR_BkXVf1LDaLI5M5TeqAFn_DI"
)
```

```python

vectorstore = Chroma.from_documents(
    documents = splits,
    embedding = gemini_embeddings,
)
```

```python
retriever = vectorstore.as_retriever()
```

```python
retriever
```

```python
system_prompt = (
    "you are an assistant for an question answering task."
    "use the following pieces of retrieved context to answer the questions"
    "if you don't know the answer say that you don't know."
    "use three sentences maximum and keep the answer concise"
    "\n\n"
    "{context}"
)
```

```python
chat_prompt = ChatPromptTemplate.from_messages(
    [
        ("system", system_prompt),
        ("human", "{input}")
    ]
)
```

```python
question_answering_chain = create_stuff_documents_chain(model, chat_prompt)
```

```python
rag_chain = create_retrieval_chain(retriever, question_answering_chain)
```

```python
rag_chain.invoke({"input": "what is MRKL?"})
```
